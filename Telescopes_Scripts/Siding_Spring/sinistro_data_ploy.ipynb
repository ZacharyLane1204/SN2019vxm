```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib widget
```

```python
sinistro_df = pd.read_csv('../../Data/sinistro_data.csv')

new_sin_df = pd.read_csv('/Users/zgl12/Research/Sn2019vxm/extra_photom_yse.csv')

new_new_sin_df = pd.read_csv('/Users/zgl12/Python_Scripts/SN2019VXM/Inputs/2019vxm_LCOphot.csv')
```

```python
new_new_sin_df = new_new_sin_df[new_new_sin_df['filter'] == 'rp'].reset_index(drop=True)
```

```python
plt.figure()
plt.errorbar(sinistro_df['mjd'], sinistro_df['mag'], yerr=sinistro_df['e_mag'], fmt='.')
plt.errorbar(new_sin_df['MJD'], new_sin_df['Mag'], yerr=new_sin_df['d_Mag'], fmt='.')
# plt.errorbar(new_new_sin_df['mjd'], -2.5*np.log10(new_new_sin_df['flux']) + 25, yerr=new_new_sin_df['d_Mag'], fmt='.')
plt.scatter(new_new_sin_df['mjd'], -2.5*np.log10(new_new_sin_df['flux']) + 25)
plt.xlabel('MJD')
plt.ylabel('Magnitude')
plt.title('Sinistro')
plt.gca().invert_yaxis()
plt.show()
```
