```python
df = pd.read_csv('../../Data/vxm_collated_data.csv')

df['Filter'] = df['Filter'].str.replace(' ', '')
```

```python
temp_df = df[df['Filter'] == 'o'].reset_index(drop=True)
```

```python
plt.figure(figsize=[fig_width, fig_height])
plt.scatter(temp_df.MJD[(temp_df.Mag > 0) & (temp_df.MJD > 58804.08)], temp_df.Mag[(temp_df.Mag > 0) & (temp_df.MJD > 58804.08)], 
            marker='*', color='darkorange', label=r'ATLAS-$o$', 
            s = 165, zorder = 10, edgecolor = 'grey', linewidth = 0.8)
plt.gca().invert_yaxis()
plt.show()
```
