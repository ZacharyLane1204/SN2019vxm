```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from glob import glob
from tqdm import tqdm

%matplotlib widget
```

```python
def main(file):
    
    df = pd.read_csv(file)
    
    title = file.split('_')
    
    # print(title)
    
    plt.figure()
    plt.errorbar(df['MJD'], df['flux'], df['d_flux'], fmt = '.')
    plt.title(title[-3] + ' ' + title[-2])
    plt.show()
```

```python
savepath = '/Users/zgl12/Modules/SN2019vxm/Telescopes_Scripts/TESS/Control_Curves/Detrended_Data/'

files = sorted(glob(savepath + '*18*psf*.csv'))

# print(files)

for file in tqdm(files, desc = 'Detrending...'):
    main(file)
    # break
```
