```python
import numpy as np
from astropy.cosmology import FlatLambdaCDM
import astropy.units as u
```

```python
c = 2.99792458e10  # speed of light in cm/s
t = 7.4240
progenitor_radius = c * t
kappa_T = 0.34  # cm^2/g, electron scattering opacity for solar metallicity
v0 = c/3  # initial velocity in cm/s
```

```python
cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
dL = cosmo.luminosity_distance(0.019)
dL = dL.to(u.cm).value
dL
```

```python
fluence = 5.0855e-07 # in erg/cm^2
observed_energy = 4*np.pi*dL**2 * fluence
observed_energy
```

```python
theoretical_energy = 8*np.pi*progenitor_radius**2 * v0 * c / kappa_T

theoretical_energy
```

```python
observed_energy/theoretical_energy
```
