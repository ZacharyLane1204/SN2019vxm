```python
import numpy as np
```

```python
def sample_param(mean, err_l, err_u, n=50000, allow_negative=False):
    samples = np.where(np.random.random(n) < 0.5,
                       np.random.normal(mean, err_l, size=n),
                       np.random.normal(mean, err_u, size=n))
    if not allow_negative:
        samples = np.clip(samples, 0, None)
    return samples
```

```python
n = 10000000

A_V_MW = 0.2760
d_A_V_MW = 0.0090

A_V_Host = 0.75
d_A_V_Host_u = 0.37
d_A_V_Host_l = 0.28

host_density_lit = 22.1
d_host_density_lit = 0.90

host_density_lc = 5.75
d_host_density_lc_u = 0.70
d_host_density_lc_l = 0.63
```

```python
A_V_MW_samples = d_A_V_MW*np.random.randn(n) + A_V_MW
A_V_Host_samples = sample_param(A_V_Host, d_A_V_Host_l, d_A_V_Host_u, n=n, allow_negative=False)
host_density_lit_samples = d_host_density_lit*np.random.randn(n) + host_density_lit
host_density_lc_samples = sample_param(host_density_lc, d_host_density_lc_l, d_host_density_lc_u, n=n, allow_negative=False)
AV_lc_samples = host_density_lc_samples/host_density_lit_samples


AV_samples = A_V_MW_samples + A_V_Host_samples + AV_lc_samples
np.nanpercentile(AV_samples, 50), np.nanpercentile(AV_samples, 50) - np.nanpercentile(AV_samples, 16), np.nanpercentile(AV_samples, 84) - np.nanpercentile(AV_samples, 50)
```
