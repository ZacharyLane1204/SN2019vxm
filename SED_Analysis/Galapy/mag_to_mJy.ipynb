```python
import numpy as np
```

```python
def mag_to_flux(mag):
    return 1e3*10**(-0.4 * (mag - 8.9))  # mJy

def mag_to_flux_and_errors(mag, mag_errs):
    flux = mag_to_flux(mag) 
    err_comp = float((0.4 * np.log(10) * mag_errs))
    
    return flux, flux * err_comp
```

```python
mag_to_flux_and_errors(17.310, 0.255)
```
